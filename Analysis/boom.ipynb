```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

df1 = pd.read_pickle("movies1.pkl")
df2 = pd.read_pickle("movies2.pkl")

df3 = pd.read_pickle("movies3.pkl")
df = pd.concat([df1,df2]).drop_duplicates().reset_index(drop=True)
GENRES = ['Action', 'Adventure', 'Animation', 'Biography', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Family', 
'Fantasy', 'Film-Noir', 'Game-Show', 'History', 'Horror', 'Music', 'Musical', 'Mystery', 'News', 'Reality-TV', 
'Romance', 'Sci-Fi', 'Sport', 'Talk-Show', 'Thriller', 'War', 'Western']
```

```python
# function to parse gross into number
def parse_gross(s):    
    try:
        if (s[-1]=="M"):
            return (float(s[1:-1]))
    except:
        return np.nan
df["Gross"] = df["Gross"].apply(parse_gross)
df["Votes"] = pd.to_numeric(df["Votes"])
```

```python
# average gross of a year
subdf = df.dropna()
subdf.groupby(['Year']).mean()
```

```python
# Scatter Plot of gross vs years
subdf = df.dropna()
x = subdf["Year"]
y = subdf["Gross"]
plt.scatter(x, y, c="g", alpha=0.5)
plt.xlabel("Years")
plt.ylabel("Gross")
plt.legend(loc=2)
plt.show()
```

```python
df_1 = df
df_y1 = df_1[np.logical_and(df_1['Year'] >= 1950, df_1['Year'] < 1960)]
df_y2 = df_1[np.logical_and(df_1['Year'] >= 1960, df_1['Year'] < 1970)]
df_y3 = df_1[np.logical_and(df_1['Year'] >= 1970, df_1['Year'] < 1980)]
df_y4 = df_1[np.logical_and(df_1['Year'] >= 1980, df_1['Year'] < 1990)]
df_y5 = df_1[np.logical_and(df_1['Year'] >= 1990, df_1['Year'] < 2000)]
df_y6 = df_1[np.logical_and(df_1['Year'] >= 2000, df_1['Year'] < 2010)]
df_y7 = df_1[df_1['Year'] >= 2010]
df_y = [df_y1, df_y2, df_y3, df_y4, df_y5, df_y6, df_y7]
# print(df1[:1])

# top gross movies
# best in every decades and their name, genres
for d in df:
    print(d.nlargest(10, "Gross"))
```

```python
import seaborn as sns; sns.set()
# a shady heatmap
sums = pd.DataFrame(columns=["Genre_"+genre for genre in GENRES])
for i in range(len(df_y)):
    vec = df_y[i].iloc[:,5:].sum()
    sums = sums.append(dict(vec), ignore_index=True)
arrayed = np.asarray(sums)
ax = sns.heatmap(arrayed)
```

```python
# extraodinary films
import nltk, re
from nltk.corpus import movie_reviews
from collections import Counter
from nltk.tokenize import word_tokenize
import random
df_sum = pd.read_pickle("mov-1.pkl")
def helper(x):
    if x < 5:
        return 0
    if x < 7:
        return np.nan
    return 1
#     if x < 7:
#         return 1
#     return 2
ind = df_sum['Rating'].apply(helper)
df_sum['ind'] = ind.values
df_sum = df_sum.drop("Gross", axis = 1)
df_sum = df_sum.dropna()
stratified = df_sum.groupby("ind")
res = []

documents = list(zip(df_sum.Summary, df_sum.ind))
random.shuffle(documents)
all_words = list(filter(lambda x: len(x) >1, (re.sub(r'[^\w\s]','',"".join(df_sum["Summary"])).lower()).split()))
all_words = nltk.FreqDist(all_words)
# print(all_words.most_common(50))
word_features = list(all_words.keys())[:3000]

def find_features(document):
    words = set(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)
    return features

featuresets = [(find_features(rev), category) for (rev, category) in documents]
training_set = featuresets[:8000]
classifier = nltk.NaiveBayesClassifier.train(training_set)



# worst ones
# stratified.count()
```
